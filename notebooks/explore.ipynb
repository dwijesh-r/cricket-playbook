```python
import duckdb
conn = duckdb.connect('data/cricket_playbook.duckdb')
print('Connected! Tables available:')
conn.sql('SHOW TABLES').show()
```

## Top Run Scorers

```python
conn.sql("""
SELECT player_name, runs, balls_faced, strike_rate, batting_average, fifties, hundreds
FROM analytics_top_run_scorers 
LIMIT 15
""").show()
```

## Top Wicket Takers

```python
conn.sql("""
SELECT player_name, wickets, overs_bowled, economy_rate, bowling_average
FROM analytics_top_wicket_takers 
LIMIT 15
""").show()
```

## Death Over Specialists

```python
conn.sql("""
SELECT player_name, overs, wickets, economy_rate, dot_ball_pct
FROM analytics_death_over_specialists 
LIMIT 10
""").show()
```

## Powerplay Hitters

```python
conn.sql("""
SELECT player_name, innings, runs, strike_rate, boundary_pct
FROM analytics_powerplay_hitters 
LIMIT 10
""").show()
```

## Head-to-Head: Kohli vs Bumrah

```python
conn.sql("""
SELECT batter_name, bowler_name, balls, runs, dismissals, strike_rate
FROM analytics_batter_vs_bowler 
WHERE batter_name LIKE '%Kohli%' 
  AND bowler_name LIKE '%Bumrah%'
""").show()
```

## Available Tables & Views

**Dimension Tables:**
- `dim_tournament` - Tournament metadata
- `dim_team` - Team info
- `dim_venue` - Venues/stadiums
- `dim_player` - Player master (with is_wicketkeeper, primary_role)
- `dim_match` - Match metadata

**Fact Tables:**
- `fact_ball` - Ball-by-ball (2.1M rows, has match_phase)
- `fact_powerplay` - Powerplay periods
- `fact_player_match_performance` - Per-match stats

**Analytics Views:**
- `analytics_batting_career` - Career batting stats
- `analytics_batting_by_phase` - Powerplay/middle/death splits
- `analytics_bowling_career` - Career bowling stats
- `analytics_bowling_by_phase` - Phase splits for bowling
- `analytics_batter_vs_bowler` - Head-to-head matchups
- `analytics_top_run_scorers` - Leaderboard (10+ innings)
- `analytics_top_wicket_takers` - Leaderboard (10+ matches)
- `analytics_death_over_specialists` - Death bowling
- `analytics_powerplay_hitters` - PP batting
